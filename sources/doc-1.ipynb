```python
import os
import pandas as pd
from rich import print
import yaml
from collections import OrderedDict
```

# Load Files

```python
# excel file name
excel_file = "estimation of heat capacities.xlsx"
# full path to the excel file
excel_file_path = os.path.join(os.getcwd(), excel_file)
print(excel_file_path)
```

```python
# load the excel file
df_1 = pd.read_excel(excel_file_path, sheet_name="Sheet1")
df_1.head(3)
```

```python
# load the excel file
df_2 = pd.read_excel(excel_file_path, sheet_name="Sheet2")
df_2.head(3)
```

# Create YML Files

```python
# create a dictionary
# looping through the rows of the dataframe
# and creating a dictionary of the data
data = {}

for index, row in df_2.iterrows():
    # set
    data[f"group-{index+1}"] = {
        'id': row['No.'],
        'name': row['Group'],
        'a': row['a[i]'],
        'b': row['b[i]'],
        'd': row['d[i]'],
        'Tmin': row['T[min]'],
        'Tmax': row['T[max]']
    }
```

```python
# Convert DataFrame to list of OrderedDict
data_list = [OrderedDict(row) for _, row in df_2.iterrows()]
```

```python
# looping through data
for key, value in data.items():
    print(f"key: {key}, value: {value}")
    break
```

```python
# Define a custom Dumper to preserve order
class OrderedDumper(yaml.Dumper):
    def represent_dict(self, data):
        return self.represent_mapping('tag:yaml.org,2002:map', data.items())

yaml.add_representer(OrderedDict, OrderedDumper.represent_dict)
```

```python
# sheet1 file
sheet1_file = "sheet2.yml"
# full path to the sheet1 file
sheet1_file_path = os.path.join(os.getcwd(), sheet1_file)
print(sheet1_file_path)
```

```python
# convert dictionary to yml file as already saved in data
with open(sheet1_file_path, "w") as f:
    yaml.dump(data_list, f, Dumper=OrderedDumper, default_flow_style=False)
```

# Load YML Files

```python
# load the yml file
with open(sheet1_file_path, "r") as f:
    data_yml = yaml.load(f, Loader=yaml.FullLoader)
```
